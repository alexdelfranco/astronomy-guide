```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
def model(radius,opacity=None,density=None):
    X = 0
    Y = 0
    X_list = []
    Y_list = []
    counter = 0
    # opacity = 1
    # density = 1
    step_length = 1 # 1 / (opacity * density)
    while m.sqrt(X**2 + Y**2) < radius:
        angle = 2 * np.pi * np.random.random()
        X_list.append(X)
        Y_list.append(Y)
        X = X + step_length * m.cos(angle)
        Y = Y + step_length * m.sin(angle)
        counter += 1
    return(counter,X_list,Y_list)
```

```python
R = 100
counter,x,y = model(R)

fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.set_title('Step Count:' + str(counter))
ax.set_xlim(-R,R)
ax.set_ylim(-R,R)
ax.plot(x,y)
ax.plot([0,x[-1]],[0,y[-1]])
ax.add_patch(plt.Circle((0,0),R,fill=False,color='red'))

print(f'Step Count: {counter}')
```
